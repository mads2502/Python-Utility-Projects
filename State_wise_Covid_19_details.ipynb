```
import folium
import pandas as pd
from geopy.geocoders import Nominatim
import time
from pprint import pprint

```

This is a small implementation of Folium in Python for visulaising COVID-19 dataset in India staewise.
The states having more than 1000 no of cases are highlighted using red colored pointers.\n

**COVID-19 DATA VISUALISER INDIA-STATE-WISE**

```

p_map=folium.Map(location=[20.5937, 78.9629],zoom_start=5) #fixing the location and zoom for India
location={}
df = pd.DataFrame(pd.read_csv('Covid cases in India.csv')) 

for i,j,k,l in zip(df['Latitude'],df['Longitude'],df[' State/ UT'],df['Total Confirmed cases']):
  temp=[]
  temp.extend((i,j))
  location['loc']=temp
  if(l<=100):
    marker=folium.Marker(location['loc'],popup=str(k)+" "+str(l)+" \nCOVID-19 data")  #placing markers in the map
  else:
    marker=folium.Marker(location['loc'],popup=str(k)+" "+str(l)+" \nCOVID-19 data",icon=folium.Icon(color="red"))
  marker.add_to(p_map)

p_map #displaying the map
```
