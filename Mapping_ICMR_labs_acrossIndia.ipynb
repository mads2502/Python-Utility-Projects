```
import folium
```

```
import pandas as pd
```

```
p_map=folium.Map(location=[20.5937, 78.9629],zoom_start=5) #fixing the location and zoom for India
location={}
df = pd.DataFrame(pd.read_csv('ICMRTestingLabsWithCoords.csv')) 

for i,j,k,l in zip(df['latitude'],df['longitude'],df['lab'],df['type']):
  temp=[]
  temp.extend((i,j))
  location['loc']=temp
  marker=folium.Marker(location['loc'],popup=str(k)+" "+str(l)+" \nICMR laboratory")  #placing markers in the map
  marker.add_to(p_map)

p_map #displaying the map
```
